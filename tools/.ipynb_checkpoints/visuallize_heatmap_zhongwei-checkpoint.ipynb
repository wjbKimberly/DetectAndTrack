```python
#create by zhongwei
import numpy as np
import json
import os
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import cv2
import glob


labels=[
    'nose',
    'head_bottom',
    'head_top',
    'left_ear',
    'right_ear',
    'left_shoulder',
    'right_shoulder',
    'left_elbow',
    'right_elbow',
    'left_wrist',
    'right_wrist',
    'left_hip',
    'right_hip',
    'left_knee',
    'right_knee',
    'left_ankle',
    'right_ankle']

def get_test_heatmap(image_heatmap_json_file_path,image_origin_root):
    print("******************* get_test_heatmap start *******************")
    if not os.path.exists(image_heatmap_json_file_path):
        print("******* the file <"+image_heatmap_json_file_path+"> is not exists!*******\n")
    print("load file <"+image_heatmap_json_file_path+"> heatmap data!")
    
    heatmaps_all_one_image=[]
    keypoints_all_one_image=[]
    
    
    with open(image_heatmap_json_file_path) as data_file:
        data=json.load(data_file)
        
    #find the image name
    image_name=data["image_name"]
    img=mpimg.imread(image_origin_root+image_name)
    print(img.shape)
    plt.title("origin image")
    plt.imshow(img)
 
    
    #find the bbox of all people in this image
    detect_bbox=data["detect"]
    
    
    #find keypoints
    kps=data["keypoints"]
    for pidx in range(len(kps)):
        kps_one_person=kps[pidx]
        keypoints_all_one_image.append(kps_one_person)
    
    #find heatmaps
    heatmaps=data["heatmaps"]
    heatmaps_np=np.asarray(heatmaps)
    print("\nthe heatmaps shape in one image:",heatmaps_np.shape)
    for personidx in range(len(heatmaps)):
        #crop the person detected and visualize
        bbox_one=detect_bbox[personidx]
        roi_img=img[int(bbox_one[1]):int(bbox_one[3]),int(bbox_one[0]):int(bbox_one[2]),:]
        print("&&&&&&&&&& roi_img:",roi_img.shape)
        h,w,c=roi_img.shape
        figure=plt.figure()
        plt.imshow(roi_img)
        plt.title("the "+str(personidx)+"th crop person")
        
        for heatmapidx in range(len(heatmaps[personidx])):
            heatmap=heatmaps[personidx][heatmapidx]
            heatmaps_all_one_image.append(heatmap)
            roi_map = cv2.resize(np.asarray(heatmap), (w, h),interpolation=cv2.INTER_CUBIC)
            
            #visualize the heatmap for every point
#             roi_map_color = cv2.applyColorMap(cv2.resize(np.asarray(heatmap), (w, h)), cv2.COLORMAP_JET)
            figure=plt.figure()
            plt.title("the "+str(personidx)+"th person "+str(heatmapidx)+"th keypoints:%s"%labels[heatmapidx],color='red')
            plt.imshow(cv2.cvtColor(roi_img, cv2.COLOR_BGR2GRAY)*0.05+roi_map)
            
            figure=plt.figure()
            plt.imshow(roi_map)

            
    print("the numbers of heatmaps in one image: ",len(heatmaps_all_one_image))

    print("******************* get_test_heatmap end *******************")
    return image_name,heatmaps_all_one_image



    
    
image_json_file_root="/var/storage/shared/pnrsy/t-jiwan/DetectAndTrack-wjb/tools/show_results/01_R101_best_hungarian-4GPU.yaml_nms=0.500000_score=_0.050000_drop_low_conf=0.950000/image_file/"
image_origin_root="/var/storage/shared/pnrsy/t-jiwan/DetectAndTrack-wjb/"

#################################################################
# #generate heatmaps just for one image
# #input the json file name for one image
# file_name="images\\bonn_5sec\\000342_mpii\\00000064.jpg.json"
# image_name,heatmaps=get_test_heatmap(image_json_file_root+file_name,image_origin_root)
#################################################################

##################################################################
#generate all heatmaps on dataset
filenames = glob.glob(image_json_file_root + "/*.json")
print(filenames[5])
for file in filenames[:1]:
    image_name,heatmaps=get_test_heatmap(file,image_origin_root)
##################################################################
```
