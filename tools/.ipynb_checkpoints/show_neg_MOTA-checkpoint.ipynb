```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import cv2
import numpy as np
import json
import os
import re

def print_mid_mota(mid_mota_dic):
    dst_keypoints = [
    '1.right_ankle',
    '2.right_knee',
    '3.right_hip',
    '4.left_hip',
    '5.left_knee',
    '6.left_ankle',
    '7.right_wrist',
    '8.right_elbow',
    '9.right_shoulder',
    '10.left_shoulder',
    '11.left_elbow',
    '12.left_wrist',
    '13.neck',
    '14.nose',
    '15.head_top',
    '16.Total']
    header=""
    FN_str="FN:\t"
    FP_str="FP:\t"
    IDSW_str="IDSW:\t"
    for i,hi in enumerate(dst_keypoints):
        header+="%s\t"%(hi)
    for i,fni in enumerate(mid_mota_dic["FN"]):
        FN_str+="%d-%s\t"%(i+1,fni)
    for i,fpi in enumerate(mid_mota_dic["FP"]):
        FP_str+="%d-%s\t"%(i+1,fpi)
    for i,idswi in enumerate(mid_mota_dic["IDSW"]):
        IDSW_str+="%d-%s\t"%(i+1,idswi)
    print(header)
    print(FN_str)
    print(FP_str)
    print(IDSW_str)
    print("gt_num:",mid_mota_dic["gt_num"])
    print("mota_i:",mid_mota_dic["mota_i"])
    print()
    
def show_bbox(coordinate,image_path,flag=0):
    img=cv2.imread(image_path)
    plt.figure(1)
    plt.imshow(img)
    if flag==0:
        x1 = coordinate[0]
        y1 = coordinate[1]
        x2 = coordinate[2]
        y2 = coordinate[3]
    
    if flag==1:
        x1 = coordinate[4]
        y1 = coordinate[5]
        x2 = coordinate[6]
        y2 = coordinate[7]
    
    currentAxis=plt.gca()
    
    for x1_i,y1_i,x2_i,y2_i in zip(x1,y1,x2,y2):
        rect=patches.Rectangle((x1_i,y1_i),x2_i-x1_i,y2_i-y1_i,linewidth=1,edgecolor='r',facecolor='none')
        currentAxis.add_patch(rect)
    
    plt.show()

def show_points(coordinate_keypoints,image_path):
    img=cv2.imread(image_path)
    plt.figure(2)
    plt.imshow(img)
    x_points = coordinate_keypoints[:,0]
    y_points = coordinate_keypoints[:,1]
    plt.plot(x_points, y_points, 'r*')
    plt.show()
    
def show_dic_i(dic_path,optical_flag):
    tmp_dic=json.load(open(dic_path))
    image_path2="/home/data/posetrack_data/posetrack_data"+tmp_dic["image_name"][0]
    image_path2=tmp_dic["image_name"]
    image_path1=image_path2[:-12]+str(tmp_dic["frame_id"]-1).zfill(8)+".jpg"
    print(image_path1)
    print(image_path2)
    w,h=tmp_dic["size"]
    ans_scores=tmp_dic["ans_scores"]
    coordinate_detect=np.asarray(tmp_dic["detect"] ).transpose([1,0])
    if optical_flag:
        pre_kps_np=np.asarray(tmp_dic["pre_keypoints"])
        pre_bbox_np=np.asarray(tmp_dic["pre_bbox"])
        coordinate_prebbox=np.asarray(tmp_dic["pre_bbox"]).transpose([1,0])
        coordinate_prekeypoints=np.asarray(tmp_dic["pre_keypoints"])
        coordinate_flow=np.asarray(tmp_dic["optical_bbox"]).transpose([1,0])
        coordinate_nms=np.asarray(tmp_dic["nms_ans"]).transpose([1,0])
        
        show_bbox(coordinate_prebbox,image_path1,0)
        show_points(coordinate_prekeypoints,image_path1)
        show_bbox(coordinate_flow,image_path2,0)
    show_bbox(coordinate_detect,image_path2,0)
    if optical_flag:
        show_bbox(coordinate_nms,image_path2,0)
    #show some cases with high conf
    print_bbox_ind=[]
    for i, scores in enumerate(ans_scores):
        if scores>0.9:
            print(scores)
            if optical_flag:
                print(tmp_dic["nms_ans"][i])
            else:
                print(tmp_dic["detect"][i])
                
root_path="show_results/01_R101_best_hungarian-4GPU-small.yaml_nms=0.500000_score=_0.850000/"
# root_path="show_results/01_R101_best_hungarian-4GPU-small-optical.yaml_optical_choice=0_nms=0.500000_score=_0.050000/"

mid_mota_path=root_path+"mid_mota/"

mota_file_list=os.listdir(mid_mota_path)
mota_file_list=[pi for pi in mota_file_list if ".jpg" in pi]
image_file_list=os.listdir(root_path)
image_file_list=[pi for pi in image_file_list if ".jpg" in pi]

mota_file_list=sorted(mota_file_list)
image_file_list=sorted(image_file_list)

optical_flag=False
if "optical" in root_path:
    optical_flag=True
    image_file_list=image_file_list[1:]
    mota_file_list=mota_file_list[1:]

#show mota frame by frame
# for path_i in mota_file_list:
#     optical_flag=False
#     if "optical" in root_path:
#         optical_flag=True
#     image_file_path_i=path_i[16:]
#     image_file_path_i=root_path+image_file_path_i
#     path_i=mid_mota_path+path_i
#     mid_mota_dic=json.load(open(path_i))
#     if True or mid_mota_dic["mota_i"]<0:
#         print("frame:%s"%path_i)
#         print_mid_mota(mid_mota_dic)
#         show_dic_i(image_file_path_i,optical_flag)
#         print()
#         print()

# show images frame by frame
for path_i in image_file_list[:50]:
    optical_flag=False
    if "optical" in root_path:
        optical_flag=True
    show_dic_i(root_path+path_i,optical_flag)
    print()
    print()
```
