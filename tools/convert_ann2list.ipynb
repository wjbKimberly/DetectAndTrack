```python
input_file_root="../../data/posetrack/posetrack_data/annotations/"
output_root="../lib/datasets/lists/PoseTrack/v2.0/"
image_root_path="../../data/posetrack/"

import json
import os
import cv2

data_split=["train","val","test"]

for split_i in data_split:
    data_final={
        "images":[],
        "categories":[],
        "annotations":[]
    }
    input_file_path_i=input_file_root+split_i
    output_file_path_i=output_root+"posetrack_%s.json"%split_i
    file_list=os.listdir(input_file_path_i)
    for index,vi_name in enumerate(file_list):
        file_i=json.load(open(input_file_path_i+"/"+vi_name))
        
        for image_i in file_i["images"]:
#             print(image_root_path+image_i["file_name"])

            from PIL import Image
            im = Image.open(image_root_path+image_i["file_name"])
            w,h = im.size
            
            image_i["width"]=w
            image_i["height"]=h
            image_i["original_file_name"]=image_i["file_name"]
        data_final["images"].extend(file_i["images"])
        
        if split_i=="test":
            data_final["annotations"]=[]
        else:
            for image_i in file_i["annotations"]:
                image_i["head_box"]=image_i["bbox_head"]
                image_i["num_keypoints"]=17
            data_final["annotations"].extend(file_i["annotations"])
        data_final["categories"]=file_i["categories"]
        print("%d/%d"%(index,len(file_list) ),vi_name,"Finished!")
    print(split_i,"Finished!")
    f=open(output_file_path_i,"w")
    f.write(json.dumps(data_final))
    f.flush()
    f.close()

```
